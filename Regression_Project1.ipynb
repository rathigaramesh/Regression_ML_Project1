```python
#PANDAS is consider as Class/library used for Data
import pandas as pd
#py -> variable
```

```python
dataset=pd.read_csv("insurance_pre.csv") #make sure file under coresponding folder
# read_csv -> pandas function
```

```python
dataset = pd.get_dummies(dataset, columns=['GENDER', 'SMOKER'], dtype=int)
```

```python
input_independent = dataset[['AGE', 'BMI', 'CHILDREN', 'GENDER_male', 'SMOKER_yes']]
```

```python
input_independent
```

```python
output_dependent =dataset[['CHARGES']]
```

```python
output_dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(input_independent, output_dependent, test_size=0.30,random_state=0)
```

```python
# from sklearn.linear_model import LinearRegression
# regressor=LinearRegression(fit_intercept=True, copy_X=True, n_jobs=None, positive=False)
# regressor.fit(X_train,y_train)

# from sklearn.svm import SVR
# regressor=SVR(kernel='sigmoid')
# regressor.fit(X_train,y_train)

# from sklearn.svm import SVR
# regressor = SVR(kernel='linear', C=100)
# regressor.fit(X_train, y_train)

# from sklearn.tree import DecisionTreeRegressor
# regressor = DecisionTreeRegressor(random_state=0)
# regressor.fit(X_train, y_train)

# from sklearn.ensemble import RandomForestRegressor
# regressor = RandomForestRegressor(n_estimators=100, random_state=42)
# regressor.fit(X_train, y_train)

# from sklearn.ensemble import RandomForestRegressor
# regressor = RandomForestRegressor()
# regressor.fit(X_train, y_train)

# from sklearn.tree import DecisionTreeRegressor
# regressor=DecisionTreeRegressor()
# regressor=regressor.fit(X_train,y_train) 

# from sklearn.ensemble import ExtraTreesRegressor
# regressor = ExtraTreesRegressor(n_estimators=100, random_state=42)
# regressor.fit(X_train, y_train)

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=200,  min_samples_split=5,  random_state=42 )
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
import pickle
filename="Final_insurance_pre_model.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("Final_insurance_pre_model.sav",'rb'))
```
